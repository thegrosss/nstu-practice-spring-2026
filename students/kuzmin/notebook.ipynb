## Mathplotlib

```python
import matplotlib.pyplot as plt

x = np.linspace(0.0, 10.0, 5000)  # Старт конец точки
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.grid()
plt.legend()
```
